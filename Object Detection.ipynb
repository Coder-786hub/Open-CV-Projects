```python
import cv2
import matplotlib.pyplot as plt
```

# Vinit 

```python
img = cv2.imread("Image/Vinit.jpg")
img = cv2.resize(img, (300, 300))
```

```python
cv2.imshow("Vinit", img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
face_cap = cv2.CascadeClassifier(r"C:\Python312\Lib\site-packages\cv2\data\haarcascade_frontalface_default.xml")
```

```python
found = face_cap.detectMultiScale(img_gray, minSize = (20, 20))
```

```python
amount_found = len(found)
```

```python
if amount_found != 0:
    for (x, y, w, h) in found:
        cv2.rectangle(img_rgb, (x, y), (x+w, y+h), (255, 0, 0), 5)

plt.subplot(1,1,1)
plt.imshow(img_rgb)
plt.show()
```

# Aftab

```python
aftab = cv2.imread("Image/aftab-Developer.jpg")
aftab = cv2.resize(aftab, (400, 400))
cv2.imshow("Aftab", aftab)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
aftab_gray = cv2.cvtColor(aftab, cv2.COLOR_BGR2GRAY)
aftab_rgb = cv2.cvtColor(aftab, cv2.COLOR_BGR2RGB)
```

```python
face_cap = cv2.CascadeClassifier(r"C:\Python312\Lib\site-packages\cv2\data\haarcascade_frontalface_default.xml")
```

```python
aftab_found = face_cap.detectMultiScale(aftab_gray, minSize = (20, 20))
```

```python
amount_aftab_found = len(aftab_found)
```

```python
if amount_aftab_found != 0:
    for (x,y,w,h) in aftab_found:
        cv2.rectangle(aftab_rgb,pt1 = (x,y), pt2 = (x+w, y+h), color = (0, 255, 0),thickness = 5)

plt.subplot(1,1,1)
plt.imshow(aftab_rgb)
plt.show()
        
```

# Stop Sign Edge Detect 

```python
stop = cv2.imread("Image/stop sign.jpeg")
stop = cv2.resize(stop, (400, 400))
cv2.imshow("Stop Sign", stop)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
stop_gray = cv2.cvtColor(stop, cv2.COLOR_BGR2GRAY)
stop_rgb = cv2.cvtColor(stop, cv2.COLOR_BGR2RGB)
```

```python
edge_detect = cv2.CascadeClassifier("stop_sign_classifier_2.xml")
```

```python
found = edge_detect.detectMultiScale(stop_gray, minSize = (20, 20))
```

```python
amount_found = len(found)
if amount_found != 0:
    for (x, y, w, h) in found:
        cv2.rectangle(stop_rgb, pt1 = (x, y), pt2 = (x+w, y+h), color = (0, 0, 255), thickness = 5)

plt.subplot(1,1,1)
plt.imshow(stop_rgb)
plt.show()
```
